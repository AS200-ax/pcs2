```python
def par(A):
    B = []
    B.append(A[0])

    for i in range(1,len(A)):
        if A[i]==A[0]:
            B.insert(0, A[i])

    for i in range(len(A)):
        if A[i] < A[0]:
            B.insert(0, A[i])
        elif A[i] > A[0]:
            B.append(A[i])
    return(B)
```

```python
data = open('rosalind_par3.txt', 'r')
data = data.readlines()
A = list(map(int, data[1].split()))
print(*par(A))
```
