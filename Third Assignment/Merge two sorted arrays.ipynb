```python
def merge(left, right):
    result = []
    while len(left) > 0 and len(right) > 0:
        if left[0] <= right[0]:
            result.append(left.pop(0))
        else:
            result.append(right.pop(0))
    
    while len(left) > 0:
        result.append(left.pop(0))
    
    while len(right) > 0:
        result.append(right.pop(0))
    
    return result

data = open('rosalind_mer.txt', 'r')
data = data.readlines()
A = list(map(int, data[1].split()))
B = list(map(int, data[3].split()))
print(*merge(A, B))
```
