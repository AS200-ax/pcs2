```python
data = open('rosalind_bins(3).txt')
d1=int(data.readline())
d2=int(data.readline())
r3=data.readline()
r4=data.readline()
ar1=r3.split(" ")
ar2=r4.split(" ")

a1 =[]
for i in ar1:
    a1.append(int(i))

a2 =[]
for i in ar2:
    a2.append(int(i))
    
def bins(A, M):
    res = []
    A.insert(0,0)
    for m in range(len(M)):
        if M[m] in A:
            res.append(A.index(M[m]))
        else: res.append(-1)
    return res


result= bins(a1,a2)
for i in result:
    print(str(i)+" ", end="")
```
