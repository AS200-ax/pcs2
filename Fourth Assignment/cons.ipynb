```python
def cons(dataset):
    #profile=np.ones((4,len(dataset,2)))
    profile=[[],[],[],[]]
    for i in range(len(dataset[0])):
        As=0
        Cs=0
        Gs=0
        Ts=0
        for j in range(len(dataset)):
            l= dataset[j]
            if l[i]=='A':
                As+=1
            if l[i]=='C':
                Cs+=1
            if l[i]=='G':
                Gs+=1
            if l[i]=='T':
                Ts+=1
        profile[0].append(As)
        profile[1].append(Cs)
        profile[2].append(Gs)
        profile[3].append(Ts)
        
    consensus=''
    for i in range(len(profile[0])):
        b1=profile[0][i]
        b2=profile[1][i]
        b3=profile[2][i]
        b4=profile[3][i]
        if max(b1,b2,b3,b4)==b1:
            consensus+='A'
        elif max(b1,b2,b3,b4)==b2:
            consensus+='C'
        elif max(b1,b2,b3,b4)==b3:
            consensus+='G'
        elif max(b1,b2,b3,b4)==b4:
            consensus+='T'
    print(consensus)
    print('A:', *profile[0])
    print('C:', *profile[1])
    print('G:', *profile[2])
    print('T:', *profile[3])
    
data=open('rosalind_cons.txt','r')
data=data.read().split('>')[1:]
dataset=[]
for i in range(len(data)):
    row=data[i].split('\n',1)
    dataset.append(row[1].replace('\n',''))
cons(dataset)
```
