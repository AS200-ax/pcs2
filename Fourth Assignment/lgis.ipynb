```python
def lgis(perm):
    incr = []
    decr = []
    for (m,n) in enumerate(perm):
        incr.append(max([incr[i] for (i, j) in enumerate(perm[:m]) if j < n] or [[]], key=len) + [n])
    for (m,n) in enumerate(perm):
        decr.append(max([decr[i] for (i, j) in enumerate(perm[:m]) if j > n] or [[]], key=len) + [n])
    print(*max(incr, key=len))
    print(*max(decr, key=len))
    
data=open('rosalind_lgis.txt','r')
data=data.readlines()
length=int(data[0])
permutation=list(map(int, data[1].split()))
lgis(permutation)
```
