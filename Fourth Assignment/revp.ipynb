```python
def comp(ssdna):
    g=''
    for i in ssdna:
        if i=="A":
            g += "T"
        if i=="T":
            g += "A"
        if i=="G":
            g += "C"
        if i=="C":
            g += "G"
    return g

def revp(dna53):
    d=[]
    dna35=comp(dna53)
    
    for j in range(4,13,2):                
        for i in range(len(dna53)):
            if i+j<len(dna53)+1:
                x=dna53[i:(i+j)]
                y=dna35[i:(i+j)][::-1]
                if x==y:
                    d.append([i+1,j])
         
    for i in range(len(d)):
        print (*d[i])

data=open('rosalind_revp.txt','r').read().split('\n',1)
dna53=data[1].replace('\n','')
revp(dna53)
```
