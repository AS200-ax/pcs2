```python
from Bio import SeqIO

def pdst(strings):
    for x in strings:
        for y in strings:
            n=0
            for i in range(len(x)):
                if x[i]!=y[i]:
                    n+=1
            print(n/len(x),end=' ')
        print('\n',end='')

strings=[]
for data in SeqIO.parse('rosalind_pdst.txt', 'fasta'):
    strings.append(data.seq)
pdst(strings)
```
